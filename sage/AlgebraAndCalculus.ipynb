```python
from sage.all import *

u = var("u")
diff(sin(u), u)
```

```python
# draw the sin(u) and its differential
p = plot(sin(u), (u, 0, 2 * pi), color="blue")
p += plot(diff(sin(u), u), (u, 0, 2 * pi), color="red")
p.show()
```
