## Define a Python function

These functions can be plotted, but **not** **differentiated** or **integrated**.

## callable symbolic expression

```python
from sage.all import *

x = var("x")
g = symbolic_expression(x ** Integer(2)).function(x)
print(g)
print(g.derivative())
plot(g, (x, -5, 5), legend_label=f"${latex(g)}$")
```

##  Accidental evaluation

```python
def sigmoid(x):
    if x < 0:
        return 0
    if x >= 0:
        return x


plot(sigmoid, (x, -5, 5), legend_label=f"sigmoid")
```

```python
def h(x):
    if x < Integer(2):
        return Integer(0)
    else:
        return x - Integer(2)


plot(h, (x, -0, 4), legend_label=f"h(x)")
```

```python
from sage.all import *

x = var("x")
f = symbolic_expression(x).function(x)
g = f.derivative()
g.variables()  # the variables present in g
# ()
g.arguments()  # the arguments which can be plugged into g
# (x,)
f = x
h = f.derivative()
h.variables()
# ()
h.arguments()
```
