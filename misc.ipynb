```python
# Import necessary libraries
from sage.all import *

# Generate a list of prime numbers between 1 and 100
primes_list = [p for p in range(1, 101) if is_prime(p)]

# Create a line plot of the prime numbers
line_plot = line([(i, p) for i, p in enumerate(primes_list)], marker="o")

print("nobody can formula for prime numbers sequences!!!")
# Show the plot
line_plot.show()
```
